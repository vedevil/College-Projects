 # Automatic speech recognition #

### 1. KNN algorithm to compute the performance of vowel classification

```python
from mat4py import loadmat    # library used fot .mat file conversion
import numpy as np            # library used for basic 
data = loadmat('data.mat')
```

```python
map={0:'a',1:'e',2:'i',3:'u'}
```

```python
training_set=[]
labels=[]
for i in range(4):
    l=len(data['vowels'][i]['training'])
    labels.extend([map[i]]*l)
    for j in range(l):
        l=list(data['vowels'][i]['training'][j])
        training_set.append(l)
```

training_set=np.array(training_set)
labels=np.array(labels)

### Split 1: Training 67% Testing 33%

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(training_set, labels, train_size=0.67)
```

```python
X_train=np.array(X_train)
y_train=np.array(y_train)
y_test=np.array(y_test)
```

```python
import knn
```

```python
y_pred=[]
for x in X_test:
    a=[]
    a.append(x)
    y=knn.predict(X_train,y_train,a,5)
    y_pred.extend(y)
    print(len(y_pred)/len(X_test)*100,end='')
    print('% done...')
    
```

#### Accuracy

```python
import sklearn.metrics
acc = sklearn.metrics.accuracy_score(y_test, y_pred)
print(acc)
```

#### Precision

```python
precision = sklearn.metrics.precision_score(y_test, y_pred,average=None)
print(precision)
```

#### Recall

```python
recall = sklearn.metrics.recall_score(y_test, y_pred, average=None)
print(recall)
```

#### F1 score

```python
f= sklearn.metrics.f1_score(y_test, y_pred, average=None)
print(f)
```
