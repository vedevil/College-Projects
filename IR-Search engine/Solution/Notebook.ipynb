```python
# importing libraries important for web crawling
import requests
import html5lib
from bs4 import BeautifulSoup
import csv
from googlesearch import search
import time
```

```python
queries=["Computer Science","IITs in India","Cities of Chattisgarh"]
i=0
j=0

# # For each query we will do the samework. So we've used the for loop here.
data="" #We've used data here give url and contents
for i in range(3):
    cr=0
    query=queries[i]
    #web crawling part
    urls=search(query,num=50,start=1, stop=40, pause=2)
    for url in urls: 
        respons=0
        try:
            respons = requests.get(url,timeout=10) #Response for the current URL
        except:
            None
        if(respons==0):
            continue
        soup = BeautifulSoup(respons.content, 'html5lib')
        paras=soup.find_all('p')  #Finding all the paragraph within the response
        para = "" #Initializing a empty paragraph to dump contents of 2 para at once.
        count=0
        for p in paras:   # Loop though each paragraph in response and append the contnet of that paragraph to  "para" so that we can save it as a continue paragraph
            para += p.get_text()
            if(p.get_text()==""):
                continue
            else:
                count+=1
            if(count==2):
                break
        if(para != ""):  #Perfom writing to file onlw when there is something in para
            data=data+"URL:"+url+"\n\n"
            data=data+"Content:\n"+para+"\n\n"
        filename="d{0}_q{1}".format(j+1,i+1)
        with open(filename, mode='wt', encoding='utf-8') as file:  #Wtite "para" to a txt file whose filename is a domain name of current URL
            file.write(para)
        j+=1
        if(j==20): #getting 20 files
            i=i+1
            j=0 
            break
    i+=1
    j=0
```

```python
import re  #importing regex module
Inverted_index = dict()
for k in range(3):
    for m in range(20):
        filename="d{0}_q{1}".format(m+1,k+1) #to read all the doc for our result
        with open(filename,'r') as f:
            for line in f:
                fline = re.sub(r'[^\w\s]', '', line).lower() #removing punctuation marks
                for string in fline.split():
                    if string not in Inverted_index: #adding each term one by one
                        l=[]
                        l.append(filename)
                        Inverted_index.__setitem__(string,l)
                    else:
                        if filename in Inverted_index[string]:
                            continue
                        else:
                            Inverted_index[string].append(filename)
                
                    
Inverted_index #inverted matrix
```

```python
#defing function to get and between two ists
#approach is same as discussed in class
def and_postings(posting1, posting2):
    p1 = 0
    p2 = 0
    result = list()
    while p1 < len(posting1) and p2 < len(posting2):
        if posting1[p1] == posting2[p2]:
            result.append(posting1[p1])
            p1 += 1
            p2 += 1
        elif posting1[p1] > posting2[p2]:
            p2 += 1
        else:
            p1 += 1
    return result
```

### Solution 3

```python
class Node:
  def __init__(self, ind = None ):
    self.ind = ind
    self.next = None

class SkipList:
  def __init__(self):
    self.head = None
```

```python
import math
def makeSkipList(p):
    q = math.ceil(math.sqrt(p))
    count = 1
    skiplist = SkipList()
    skiplist.head = Node(0)
    prev = skiplist.head
    for i in range(1, p):
        if count == q:
            prev.next = Node(i)
            count = 0
            prev = prev.next
        count += 1
    return skiplist.head
```

```python
def skipintersect(l1 :list, l2 :list, node1 :Node, node2 :Node):
    i = 0
    j = 0
    res = []
    while i < len(l1) and j < len(l2):
        if l1[i] == l2[j]:
            res.append(l1[i])
            i += 1
            j += 1
        elif l1[i] < l2[j]:
            if node1 is not None and node1.ind == i:
                while node1.next is not None and l1[node1.next.ind] < l2[j]:
                    i = node1.next.ind
                    node1 = node1.next
            i += 1
        else:
            if node2 is not None and node2.ind == j:
                while node2.next is not None and l2[node2.next.ind] < l1[i]:
                    j = node2.next.ind
                    node2 = node2.next
            j += 1
        while node1 is not None and i > node1.ind:
            node1 = node1.next
        while node2 is not None and j > node2.ind:
            node2 = node2.next
    return res
```

```python
skip_terms = dict()
for term in Inverted_index.keys():
    skip_terms[term] = makeSkipList(len(Inverted_index[term]))
#skip_terms
```

```python
print('skiplist for bhilai',skip_terms['bhilai'])
print('Skiplist for iit',skip_terms['iit'])
print('skiplist for computer',skip_terms['computer'])
```
